***Convolution Layer (Conv)***

---
- Convolution Layer là một loại lớp trong mạng neural (neural network) được sử dụng trong deep learning.

- Nó thực hiện một tác vụ biến đổi hình ảnh từ một hình dạng này sang hình dạng khác bằng cách tính toán tích chập giữa các kernal (bộ lọc) và các phần tử của hình ảnh. 

- Kết quả của tích chập đó sẽ được sử dụng để xác định những đặc trưng quan trọng của hình ảnh.

- Convolution Layer là một trong những tầng cơ bản trong một mạng neural để nhận diện đặc trưng và đạt đến kết quả tốt trong các bài toán học số học.
![](https://stanford.edu/~shervine/teaching/cs-230/illustrations/convolution-layer-a.png?1c517e00cb8d709baf32fc3d39ebae67)


***Pooling(POOL)***

---

- Pooling trong mạng neural (neural network) là một tầng cấp độ trong Convolutional Neural Network (CNN). Nó có chức năng giảm chiều dữ liệu và tăng tốc độ huấn luyện mạng.

- Pooling hoạt động bằng cách chia tấm ảnh thành nhiều vùng nhỏ hơn và chọn một giá trị đại diện cho mỗi vùng, chẳng hạn như giá trị lớn nhất hoặc trung bình. Sau đó, sẽ gôm 1 tập các điểm ảnh lại với nhau và cho đầu ra thành 1 tập nhỏ hơn. Điều này giảm chiều dữ liệu, giảm số lượng các tham số cần huấn luyện và tránh overfitting. 
![](https://i.imgur.com/S6DhCs3.png)
---
> Ví dụ, max pooling sẽ lựa chọn giá trị lớn nhất trong mỗi vùng và average pooling sẽ tính trung bình các giá trị trong mỗi vùng.

⇒ *Pooling là một tầng quan trọng trong một mạng neural để xác định đặc trưng và đạt đến kết quả tốt trong các bài toán học số học.*
![](https://i.imgur.com/fcaRYS6.png)
***Fully Connected (FC)***

---

- Fully Connected (FC) layer trong Convolutional Neural Network (CNN) là một tầng cấp độ cuối cùng trong một mạng neural. Nó là một tầng fully connected đầy đủ, tức là tất cả các đơn vị trong tầng trước đều được kết nối với tất cả các đơn vị trong tầng này.

- FC layer có chức năng dự đoán nhãn cho một tấm ảnh dựa trên các đặc trưng được tìm thấy trong các tầng trước. Đặc trưng được tổng hợp từ các tầng Convolution và Pooling và sau đó được sử dụng để dự đoán nhãn cho tấm ảnh.

- Tầng FC cần huấn luyện các tham số để tối ưu dự đoán của nó. Khi huấn luyện xong, mạng neural sẽ sử dụng tầng FC để dự đoán nhãn cho tấm ảnh mới.

### ***Giới thiệu***
---

- ResNet (Residual Network) là loại mạng nơ-ron tích chập (CNN) được giới thiệu năm 2015 bởi những nhà nghiên cứu của Microsoft. Nó được thiết kế để giải quyết vấn đề của các gradient biến mất trong các mạng rất sâu, trong đó chính xác của mô hình giảm xuống khi số lớp tăng lên.

- ResNets đã trở nên phổ biến trong các tác vụ thị giác máy tính như phân loại hình ảnh và phát hiện đối tượng, đồng thời đã đạt được kết quả tiên tiến nhất trên các bộ dữ liệu điểm chuẩn khác nhau. Chúng cũng đã được sử dụng trong nhiều tác vụ khác như phân loại video, phân đoạn ngữ nghĩa và chuyển kiểu.
---


### ***Tại sao lại xuất hiện mạng ResNet***
---
> Mạng ResNet là một mạng CNN được thiết kế để làm việc với hàng trăm lớp. Một vấn đề xảy ra khi xây dựng mạng CNN với nhiều lớp chập sẽ xảy ra hiện tượng Vanishing Gradient dẫn tới quá trình học tập không tốt.
---
#### ****Vanishing Gradient****
---
> Trước hết thì Backpropagation Algorithm là một kỹ thuật thường được sử dụng trong quá trình tranining. Ý tưởng chung của thuật toán lá sẽ đi từ output layer đến input layer và tính toán gradient của cost function tương ứng cho từng parameter (weight) của mạng. Gradient Descent sau đó được sử dụng để cập nhật các parameter đó.

![](https://images.viblo.asia/2c64f987-0660-4f19-ba0e-caf75d6ed1c2.png)

######*VGG-19 là một mô hình CNN sử dụng kernel 3x3 trên toàn bộ mạng, VGG-19 cũng đã giành được ILSVRC năm 2014.*

> Toàn bộ quá trình trên sẽ được lặp đi lặp lại cho tới khi mà các parameter của network được hội tụ. Thông thường chúng ta sẽ có một hyperparametr (số Epoch - số lần mà traninig set được duyệt qua một lần và weights được cập nhật) định nghĩa cho số lượng vòng lặp để thực hiện quá trình này. Nếu số lượng vòng lặp quá nhỏ thì ta gặp phải trường hợp mạng có thể sẽ không cho ra kết quả tốt và ngược lại thời gian tranining sẽ lâu nếu số lượng vòng lặp quá lớn.

![](https://images.viblo.asia/4f9fd490-ebe3-4346-960d-0fa2a99eec76.png)


> Tuy nhiên, trong thực tế Gradients thường sẽ có giá trị nhỏ dần khi đi xuống các layer thấp hơn. Dẫn đến kết quả là các cập nhật thực hiện bởi Gradients Descent không làm thay đổi nhiều weights của các layer đó và làm chúng không thể hội tụ và mạng sẽ không thu được kết quả tốt. Hiện tượng như vậy gọi là Vanishing Gradients.

### ***Kiến trúc***

---
> Giải pháp mà ResNet đưa ra là sử dụng kết nối "tắt" đồng nhất để xuyên qua một hay nhiều lớp. Một khối như vậy được gọi là một Residual Block, như trong hình:

![](https://images.viblo.asia/full/9fc720ba-45e4-4d2f-8b58-5926350bc075.png)
######*ResNet sử dụng các kết nối tắt ( kết nối trực tiếp đầu vào của lớp (n) với (n+x) được hiển thị dạng mũi tên cong. Qua mô hình nó chứng minh được có thể cải thiện hiệu suất trong quá trình training model khi mô hình có hơn 20 lớp.*

Ảnh bên trên hiển thị khối dư được sử dụng trong mạng. Xuất hiện một mũi tên cong xuất phát từ đầu và kết thúc tại cuối khối dư. 

Hay nói cách khác là sẽ bổ sung Input X vào đầu ra của layer, hay chính là phép cộng mà ta thấy trong hình minh họa, việc này sẽ chống lại việc đạo hàm bằng 0, do vẫn còn cộng thêm X. Với H(x) là giá trị dự đoán, F(x) là giá trị thật (nhãn), chúng ta muốn H(x) bằng hoặc xấp xỉ F(x). 

- Việc F(x) có được từ x như sau:
$X → \text{weight}_1 → ReLU → \text{weight}_2$


- Giá trị H(x) có được bằng cách:
$F(x) + x → ReLU$

#### *Một số biến thể phổ biến của kiến trúc ResNet:*

---
- ResNet-18: Mô hình này có 18 weight layers và tương đối shallow (nông) so với các mô hình ResNet khác.

- ResNet-34: Mô hình này có 34 weight layers và deep(sâu) hơn ResNet-18 nhưng shallow (nông) hơn ResNet-50.

- ResNet-50: Mô hình này có 50 weight layers và là một trong những mô hình ResNet được sử dụng rộng rãi nhất.

- ResNet-101: Mô hình này có 101 weight layers và deep (sâu) hơn ResNet-50.

- ResNet-152: Mô hình này có 152 weight layers và là lớp deep sâu nhất trong số các mô hình ResNet.
---

### ***Nguyên tắc hoạt động***
---

> Nguyên tắc hoạt động của mô hình ResNet dựa trên các kết nối còn lại, cho phép mô hình tìm hiểu residual connections giữa đầu vào và đầu ra mong muốn.

- Trong deep neural network, input được truyền qua multiple layers và ouput của mỗi layer được sử dụng làm input cho layer tiếp theo. Tuy nhiên, khi số lượng layer tăng lên, mô hình sẽ khó tìm hiểu và tối ưu hóa các tham số hơn, dẫn đến vanishing gradient.

- Mô hình ResNet giải quyết vấn đề này bằng cách thêm các shortcut connections giữa các layers, cho phép input được thêm trực tiếp vào ouput của layer. Điều này có nghĩa là mô hình có thể tìm hiểu residual information, đó là sự khác biệt output ra mong muốn và input, thay vì hàm ban đầu.

- Trong mô hình ResNet, mỗi layer output từ layer trước đó làm input, sau đó sử dụng một loạt các convolutional, chuẩn hóa hàng loạt và activation layers để xử lý input. Ouput của layer sau đó được thêm vào input, tạo ra residual information.

- Việc sử dụng cácresidual connections giúp có thể train extremely deep networks, với hàng trăm layer mà không gặp phải vanishing gradient. Ngoài ra, việc sử dụng các hàm ReLU activation và chuẩn hóa hàng loạt giúp cải thiện hơn nữa độ chính xác và ổn định của mô hình.

### ***Các dạng Layers***

---
![](https://pub.mdpi-res.com/applsci/applsci-09-04209/article_deploy/html/images/applsci-09-04209-g001.png?1572358128)

***Convolution Layer (Conv)***

---

- Convolutional layers là một block xây dựng cơ bản trong các mô hình thị giác máy tính. 

- Trong mô hình ResNet, các layers này được sử dụng để trích xuất các tính năng từ hình ảnh đầu vào. Hoạt động convolution liên quan đến việc áp dụng một tập hợp các bộ lọc cho input, giúp nắm bắt các patterns và relationships trong data. 
- Kích thước không gian của input thường được giảm thông qua việc sử dụng các pooling layers, điều này cũng giúp giảm độ phức tạp tính toán.
---


***Batch Normalization Layers:***

---

- Batch Normalization Layers là một kỹ thuật để chuẩn hóa activations của từng layer trong network. 
- Điều này giúp cải thiện tính ổn định và hội tụ của mô hình, cũng như giảm tác động của sự dịch chuyển đồng biến, đó là sự thay đổi trong phân phối dữ liệu đầu vào khi nó đi qua network. 

- Trong mô hình ResNet, các lớp chuẩn hóa hàng loạt được sử dụng để chuẩn hóa activations trước khi chúng được chuyển qua hàm activation.
---


***Activation Layers:***

---

- Activation Layers đưa tính phi tuyến tính vào mô hình, cho phép nó mô hình hóa các hàm phức tạp hơn. 

- Hàm ReLU activation thường được sử dụng trong các mô hình ResNet, cho ra output tối đa là 0 và input activation. 

- Điều này giúp ngăn việc activations trở nên quá nhỏ hoặc quá lớn, điều này có thể gây ra vanishing gradient.
---


***Residual Connections:***

---

- Residual connections là key innovation trong các mô hình ResNet và cho phép mô hình tìm hiểu residual information giữa input và output mong muốn. 

- Trong mô hình ResNet, đầu ra của mỗi layer được thêm vào input, tạo ra residual information. 

- Điều này cho phép huấn luyện các deep networks với số lượng lớn các layers, vì mô hình có thể tìm hiểu residual information ngay cả khi ánh xạ input-output quá phức tạp để được mô hình hóa trực tiếp.
---


***Pooling Layers:***

---
- Pooling layers được sử dụng để giảm kích thước không gian của input, giúp nắm bắt các features  quan trọng nhất và giảm độ phức tạp tính toán. 

- Max pooling là một phương pháp pooling phổ biến được sử dụng trong các mô hình ResNet, phương pháp này đưa ra giá trị tối đa trong một tập hợp các input activations. 

- Điều này giúp giảm kích thước không gian của input trong khi vẫn giữ được thông tin quan trọng nhất.
---


***Fully Connected Layers:***

---

- Fully connected layers được sử dụng ở cuối mô hình để đưa ra dự đoán cuối cùng. 

- Các lớp này thường bao gồm các lớp dense, feedforward và được sử dụng để xử lý các activations từ final convolutional layer và đưa ra dự đoán cuối cùng. 

- Trong mô hình ResNet, số lượng nơ ron trong fully connected layer thường bằng với số lượng classes trong dataset và các activations được chuyển qua hàm softmax activation để tạo ra phân phối xác suất trên các classes.
---

## ▶◀ **Sự khác nhau giữa mô hình CNN cơ bản và ResNet**

### ***Architecture (Kiến trúc)***

---
> Một mô hình CNN cơ bản là một feedforward neural network bao gồm nhiều multiple convolutional và pooling layers. Mặt khác, ResNet là một loại CNN sử dụng các residual connections giữa các layers để cho phép network truyền gradient tốt hơn trong quá trình training.
---



### ***Depth (Độ sâu)***

---
> Các mô hình ResNet thường sâu hơn các mô hình CNN cơ bản và có thể có hàng trăm layers. Điều này cho phép ResNet tìm hiểu các biểu diễn phức tạp hơn của input data.
---

### ***Skip connections (Bỏ qua các kết nối)***

---
> Các mô hình CNN cơ bản không có skip connections, trong khi các mô hình ResNet sử dụng chúng để cho phép information vượt qua multiple layers và tiếp cận các deeper layers hơn dễ dàng hơn. Điều này giúp giảm bớtvanishing gradient trong các deep neural networks.
---

### ***Training (Huấn luyện)***

---
> Các mô hình ResNet thường dễ đào tạo hơn các mô hình CNN cơ bản do các residual connections của chúng và cách chúng truyền gradient trong quá trình training.
---

### ***Performance (Hiệu suất)***

---
> Các mô hình ResNet đã được chứng minh là hoạt động tốt hơn các mô hình CNN cơ bản trong nhiều tác vụ thị giác máy tính, bao gồm phân loại hình ảnh và phát hiện đối tượng.
---

### ***Parameters (Tham số)***

---
> Các mô hình ResNet thường có nhiều parameters (tham số) hơn các mô hình CNN cơ bản do kiến trúc sâu hơn của chúng và các tham số bổ sung cần thiết cho các residual connections.

### ***ResNet (Residual Network)***

```python
#Defines a residual block in a ResNet architecture
def residual_block(x, filters, strides=1):
    # First Conv2D layer
    x_res = keras.layers.Conv2D(filters, (3, 3), strides=strides, padding='same')(x)
    # Batch normalization layer
    x_res = keras.layers.BatchNormalization()(x_res)
    # ReLU activation layer
    x_res = keras.layers.Activation('relu')(x_res)
    # Second Conv2D layer
    x_res = keras.layers.Conv2D(filters, (3, 3), strides=1, padding='same')(x_res)
    # Batch normalization layer
    x_res = keras.layers.BatchNormalization()(x_res)
    # Downsampling path
    x = keras.layers.Conv2D(filters, (1, 1), strides=strides, padding='same')(x)
    x = keras.layers.BatchNormalization()(x)
    # Add the residual connection
    x = keras.layers.add([x, x_res])
    # ReLU activation layer
    x = keras.layers.Activation('relu')(x)
    return x
```

```python
def ResNet50(input_shape=(28, 28, 1), num_classes=10):
    # Input layer
    inputs = keras.Input(shape=input_shape)
    # First Conv2D layer
    x = keras.layers.Conv2D(64, (7, 7), strides=2, padding='same')(inputs)
    # Batch normalization layer
    x = keras.layers.BatchNormalization()(x)
    # ReLU activation layer
    x = keras.layers.Activation('relu')(x)
    # Max pooling layer
    x = keras.layers.MaxPool2D((3, 3), strides=2, padding='same')(x)
    filters = 64
    for i in range(3):
        # Determine the stride for the current residual block
        strides = 1 if i == 0 else 2
        # Call the residual block function
        x = residual_block(x, filters, strides)
        # Double the number of filters
        filters *= 2
    # Global average pooling layer
    x = keras.layers.GlobalAveragePooling2D()(x)
    # Output dense layer with a softmax activation function
    x = keras.layers.Dense(num_classes, activation='softmax')(x)
    # Model definition
    ResNet50Model = keras.Model(inputs, x)
    return ResNet50Model
```

### ***Basic CNN model (Convolutional Neural Network)***

```python
fashion_mnist = tf.keras.datasets.fashion_mnist
(train_images, train_labels), (test_images, test_labels) = fashion_mnist.load_data()
# tập train_labels sẽ được đánh dấu kiểu dáng bằng số để tránh việc tiếng anh hóa một kiểu dáng
```

```python
# Build a basic CNN model
CNNmodel = tf.keras.models.Sequential([
    tf.keras.layers.Conv2D(64, (3,3), activation = 'relu',
                           input_shape = (28, 28, 1)),
    tf.keras.layers.MaxPool2D(2,2),
    tf.keras.layers.Conv2D(64, (3,3), activation = 'relu'),
    tf.keras.layers.MaxPool2D(2,2),
    tf.keras.layers.Conv2D(64, (3,3), activation = 'relu'),
    tf.keras.layers.MaxPool2D(2,2),
    tf.keras.layers.Flatten(input_shape = (28, 28)),
    tf.keras.layers.Dense(128, activation = tf.nn.relu),
    tf.keras.layers.Dense(10, activation = tf.nn.softmax)
])
```

```python
CNNmodel.summary()
```

```python
# Plot the ConvNet model
plot_model(CNNmodel, to_file='ConvNetModel.png')
```

```python
# Compile the ConvNet model
CNNmodel.compile(optimizer = tf.keras.optimizers.Adam(),
              loss = 'sparse_categorical_crossentropy',
              metrics=['accuracy'])
```
